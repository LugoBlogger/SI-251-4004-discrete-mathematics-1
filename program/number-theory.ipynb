Fast modular exponentiation

```python
def mod_exponent(b: int, n: str, m: int):

  x = 1
  power = b % m

  # -- alternative 1
  # k = len(n)
  # for i in range(0, k):
  #   a_i = int(n[k-1 - i])
  #   if a_i == 1: x = (x * power) % m
  #   power = (power * power) % m


  # -- alternative 2
  for a_i in n[::-1]:
    if int(a_i) == 1: x = (x * power) % m
    power = (power * power) % m

  return x
```

Linear congruence
$$
ax \equiv b \pmod{m}
$$

We can change it into a Diophantine equation (from the 
definition of modular arithmetic)
$$
ax - b = my
$$
where $x$ and $y$ are integers

```python
# Example of 5x \equiv 3 (mod 7)

a = 5
b = 3
m = 7

x_sol = []
y_sol = []
x_min = -20
x_max = 20
for x in range(x_min, x_max+1):
  if (a*x - b) % m == 0:
    x_sol.append(x)
    y_sol.append((a*x - b) // m)
print(x_sol)
print(y_sol)


fig, ax = plt.subplots()

ax.scatter(x_sol, y_sol)

ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_aspect("equal")
ax.set_title("$ax = b \,\,(\operatorname{mod}{m})$;"
  + " $ax - b = my$;" + f"$a={a}$; $b={b}$; $m={m}$")
ax.grid("on")

plt.show(fig)
```
