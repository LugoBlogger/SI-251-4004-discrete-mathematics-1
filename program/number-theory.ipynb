```python
def mod_exponent(b: int, n: str, m: int):

  x = 1
  power = b % m

  # -- alternative 1
  # k = len(n)
  # for i in range(0, k):
  #   a_i = int(n[k-1 - i])
  #   if a_i == 1: x = (x * power) % m
  #   power = (power * power) % m


  # -- alternative 2
  for a_i in n[::-1]:
    if int(a_i) == 1: x = (x * power) % m
    power = (power * power) % m

  return x
```
