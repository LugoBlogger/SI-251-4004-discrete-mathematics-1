```python
import json
import matplotlib.pyplot as plt
```

```python
plt.rcParams.update(plt.rcParamsDefault)
plt.rcParams.update({
  'font.size': 16,
  'grid.alpha': 0.25})
```

```python
with open("./words_dictionary.json") as fp:
  data = json.load(fp)
```

```python
all_words = list(data.keys())
all_words
```

Calculate the frequency for each letter

```python
letter_freq = {}
for word in all_words:
  for letter in word:
    if letter not in letter_freq:
      letter_freq[letter] = 1
    else:
      letter_freq[letter] += 1

letter_freq
```

```python
sorted_letter_freq = dict(
  sorted(letter_freq.items(), key=lambda item: item[1], reverse=True))
sorted_letter_freq
```

```python
fig, ax = plt.subplots(figsize=(8, 3))

ax.bar(sorted_letter_freq.keys(), sorted_letter_freq.values())

total_letters = sum(sorted_letter_freq.values())
xlocks = ax.get_xticks()
for i, v in enumerate(sorted_letter_freq.values()):
  percentage = v/total_letters * 100
  ax.text(xlocks[i] - 0.25, v + 10_000, str(round(v/1_000))+"k" + f"({percentage:.1f}%)", fontsize=12, rotation=45)

ax.grid("on")

ax.set_ylim([0, 420_000])
ax.set_ylabel("freq.")
ax.set_xlabel("letter")

plt.show(fig)
```
