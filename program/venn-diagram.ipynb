### $(A \cup B) - (A \cap B)$

$$
\begin{align*}
(A \cup B) - (A \cap B)
  &= (A \cup B) \cap \overline{(A \cap B)} \\
  &= (A \cup B) \cap (\overline{A} \cup \overline{B}) 
\end{align*}
$$

```python
fig, ax = plt.subplots(figsize=(10, 4))

width = 2  
height = 1.2 
rect = Rectangle((-width/2, -height/2), width, height, alpha=1, zorder=-3,
                  facecolor="w", edgecolor="k")
ax.add_patch(rect)

v = venn2(subsets=(1., 1., .3), set_labels=None, ax=ax, alpha=1)

set_A = v.get_patch_by_id("10")
set_A_label = v.get_label_by_id("10")
set_A.set_edgecolor("k")
set_A.set_linewidth(2)
set_A.set_facecolor("#e6e7e8ff")
set_A_label.set_text("$A$")
set_A_label.set_fontsize(24)

set_B = v.get_patch_by_id("01")
set_B_label = v.get_label_by_id("01")
set_B.set_edgecolor("k")
set_B.set_linewidth(2)
set_B.set_facecolor("#e6e7e8ff")
set_B_label.set_text("$B$")
set_B_label.set_fontsize(24)

set_AandB = v.get_patch_by_id("11")
set_AandB_label = v.get_label_by_id("11")
set_AandB.set_edgecolor("k")
set_AandB.set_linewidth(2)
set_AandB.set_facecolor("w")
set_AandB_label.set_text("")

padding = .1
ax.set_xlim([-width/2 - padding, width/2 + padding])
ax.set_ylim([-height/2 - padding, height/2 + padding])

plt.show(fig)
```

### $(\overline{A} \cap \overline{B}) \cup (A \cap B)$

$$
\begin{align*}
  \overline{((A \cup B) - (A \cap B))}
    &= \overline{(A \cup B) \cap \overline{(A \cap B)}} \\
    &= \overline{(A \cup B)} \cup \overline{\overline{(A \cap B)}} \\
    &= (\overline{A} \cap \overline{B}) \cup (A \cap B) \\
\end{align*}
$$

    &= (\overline{A} \cup \overline{B}) \cap
